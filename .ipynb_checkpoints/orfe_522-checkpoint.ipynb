```python
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (4, 3)  
plt.rcParams["axes.grid"] = True          
plt.rcParams["figure.autolayout"] = True 
import cvxpy as cp
```

```python
# Inputs of the problem
n = 2
N = 20
A = np.array([[1., 1.],
              [0., 0.95]])
b = np.array([0., 0.1])
x_0 = np.array([0, 0])
x_des = np.array([10, 0])
```

# Solution 1

We let $\texttt{CVXPY}$ convert the problem to an LP one.

```python
# Variables
X = cp.Variable((n, N+1))
u = cp.Variable((1, N))

# Constraints
constraints = [ 
    X[:, 1:] == A @ X[:, :-1] + b[:, None] @ u,
    X[:, 0] == x_0,
    X[:, -1] == x_des,
    ]

# Objective
objective = cp.sum(cp.maximum(cp.abs(u), 2*cp.abs(u)-1))

# Problem
problem = cp.Problem(cp.Minimize(objective), constraints)

# Solve
sol = problem.solve()
print("The total fuel consumption is %.2f" % sol)
```

```python
# Plot
plt.plot(u.value[0], label=r'Input Signal $u(t)$')
plt.plot(X.value[0], label=r'Position $X_{1}(t)$')
plt.plot(X.value[1], label=r'Velocity $X_{2}(t)$')
plt.xlabel('Time step')
plt.legend()
plt.show()
```

# Solution 2

We direcly apply our LP formulation. 

```python
# Variables
X = cp.Variable((n, N+1))
u = cp.Variable((1, N))
f = cp.Variable((1, N))

# Constraints
constraints = [
    u <= f,
    -u <= f,
    2*u-1 <= f,
    -2*u-1 <= f,
    X[:, 1:] == A @ X[:, :-1] + b[:, None] @ u,
    X[:, 0] == x_0,
    X[:, -1] == x_des,
]

# Objective
objective = cp.sum(f)

# Problem
problem = cp.Problem(cp.Minimize(objective), constraints)

# Solve
sol = problem.solve()
print("The total fuel consumption is %.2f" % sol)
```

```python
# Plot
plt.plot(u.value[0], label=r'Input Signal $u(t)$')
plt.plot(X.value[0], label=r'Position $X_{1}(t)$')
plt.plot(X.value[1], label=r'Velocity $X_{2}(t)$')
plt.xlabel('Time step')
plt.legend()
plt.show()
```

# The solutions are identical

```python
print(f'The input signals are identical: {np.allclose(u.value, u.value)}')
print(f'The state trajectories are identical: {np.allclose(X.value, X.value)}')
```
